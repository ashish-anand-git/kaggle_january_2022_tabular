```python
import pandas as pd

#Loading data
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
test_id=test.row_id

all_data=pd.concat([train,test])
all_data
```

```python
#converting date into usable data

all_data['date'] = pd.to_datetime(all_data['date'])
all_data['year'] = all_data['date'].dt.year
all_data['month'] = all_data['date'].dt.month
all_data['day'] = all_data['date'].dt.day
all_data['dayofweek'] = all_data['date'].dt.dayofweek
all_data['dayofmonth'] = all_data['date'].dt.days_in_month
all_data['dayofyear'] = all_data['date'].dt.dayofyear
all_data['weekday'] = all_data['date'].dt.weekday
all_data['weekofyear'] = all_data['date'].dt.isocalendar().week

all_data.drop(columns = ['num_sold', 'date', 'row_id'], inplace = True)
```

```python
#encoding catagorical value

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in ['country', 'product', 'store']:
    all_data[col] = le.fit_transform(all_data[col])
```

```python
X_train = all_data[:len(train)]
X_test = all_data[len(train):]
y_train = train['num_sold']
```

```python
#implementing random forest model
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X_train, y_train)
preds = model.predict(X_test)
```

```python
output=pd.DataFrame({'row_id': test_id, 'num_sold': preds})
output.to_csv('submission2.csv',index=False)
```
